```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([100, 500])
y_train = np.array([10, 100])
m = x_train.shape[0]
```

```python
def compute_cost(x, y, w, b):
    total_cost = 0
    for i in range(m):
        total_cost += (w * x[i] + b - y[i]) **2
    return total_cost/(2*m) 
```

```python
def compute_gradient(x, y, w, b):
    dj_dw = 0
    dj_db = 0
    for i in range(m):
        dj_dw += (w*x[i] + b - y[i])*x[i]
        dj_db += (w*x[i] + b - y[i])
    return dj_dw/m, dj_db/m    
```

```python
def gradient_descent(x, y, w, b, alpha, num_iters):
    cost_history = []
    for i in range(num_iters):
        dw, db = compute_gradient(x, y, w, b)
        w = w - alpha*dw
        b = b - alpha*db
        cost_history.append(compute_cost(x, y, w, b))

        if i %10 == 0:
            print(f"Iteration {i}: Cost = {cost_history[-1]:.4f}, w = {w:.2f}, b = {b:.2f}")

    return w, b, cost_history        
```

```python
initial_w = 0
initial_b = 0
learning_rate = 0.00001
iterations = 1000

final_w, final_b, cost_history = gradient_descent(x_train, y_train, initial_w, initial_b, learning_rate, iterations)
```

```python
def predict (w, x, b):
    return w*x + b
```

```python
predicted_price = predict(final_w, 300, final_b)
print(f"\nPrice of 300 sqft house: ${predicted_price * 1000:.2f}")
```

```python
plt.scatter(x_train, y_train, color='blue', label='Data Points')
plt.plot(x_train, predict(final_w, x_train, final_b), color='red', label='Regression Line')
plt.xlabel("Size (sq ft)")
plt.ylabel("Price ($1000s)")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid(True)
plt.show()

# Plot cost vs iteration
plt.plot(range(iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost Function Convergence")
plt.grid(True)
plt.show()
```
